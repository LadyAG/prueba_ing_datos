```python
# write query inside docstring
query_2 = """
    Select a.UserName, 
           (select count(*) from train_table where username = a.username and sentiment = 'Positive') pos_count,
           (select count(*) from train_table where username = a.username and sentiment = 'Negative') neg_count,
           (select count(*) from train_table where username = a.username and sentiment = 'Extremely Positive') extremely_pos_count,
           (select count(*) from train_table where username = a.username and sentiment = 'Extremely Negative') extremely_neg_count
    From train_table a
    Group by a.userName
"""
```
